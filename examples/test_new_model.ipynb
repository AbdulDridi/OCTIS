```python
import os
import numpy as np
os.chdir(os.path.pardir)
from dataset.dataset import Dataset
```

```python
dataset = Dataset()
dataset.load("preprocessed_datasets/newsgroup/newsgroup_lemmatized_10")
```

```python
data_corpus = [','.join(i) for i in dataset.get_corpus()]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(token_pattern=r'(?u)\b\w+\b')
X = vec.fit_transform(data_corpus)
```

```python
idx2token = {v: k for (k, v) in vec.vocabulary_.items()}
#train_data = datasets.BOWDataset(X.toarray(), idx2token)
input_size = len(idx2token.keys())
def split_bow(bow_in, n_docs):
    indices = [[w for w in bow_in[doc,:].indices] for doc in range(n_docs)]
    counts = [[c for c in bow_in[doc,:].data] for doc in range(n_docs)]
    return indices, counts
```

```python
bow_ts_h1_tokens, bow_ts_h1_count = split_bow(X, X.shape[0])
```
